```python
import pandas as pd
```

```python
dataset = pd.read_csv("50_Startups.csv")
```

```python
dataset
```

```python


dataset = pd.get_dummies(dataset,drop_first =True)
```

```python

dataset.columns
```

```python
independent = dataset[['R&D Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_New York']]
```

```python
dependent = dataset ['Profit']
```

```python

independent
```

```python
dependent
```

```python
from sklearn.model_selection import train_test_split
```

```python
 x_train,x_test,y_train,y_test =train_test_split(independent, dependent,test_size =0.30,random_state=0)
```

```python

import matplotlib.pyplot as plt
```

```python
from sklearn import tree
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
regressor = DecisionTreeRegressor(criterion ='mae',splitter = 'best',max_features = 'auto')
```

```python
#regressor = DecisionTreeRegressor(criterion ='mae',splitter = 'best',max_features = 'auto',max_depth = 10,min_impurity_decrease = 15.6,max_leaf_nodes=20,min_samples_leaf=3,min_weight_fraction_leaf=0.4,min_samples_split =12,presort = False)
```

```python
regressor.fit(x_train,y_train)
```

```python
regressor.tree_
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
tree.plot_tree(regressor)
plt.show()
```

```python
y_pred = regressor.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r_score = r2_score(y_test,y_pred)
```
